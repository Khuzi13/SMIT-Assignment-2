
## -o) Data Cleaning

```python
# o) Data Duplicacy
print('\nOverall Duplicate Data Count : ',df.duplicated().sum()) # no data duplicacy
print('\nSame Id Record Count : ',df['Row ID'].duplicated().sum()) # no repeated id record
```

```python
# o) Data Uniqueness
for i in df.columns:
    print(f'Colume Name : {i}',f'Uniqueness of {i} : {df[i].nunique()}',sep='\n')
    print(df[i].unique(),end='\n\n')
```

```python
# o) Data Skewness
for i in df.select_dtypes(include='number').columns:
    print(f'Colume Name : {i}',f'Skewness of {i} : {df[i].skew()}',sep='\n')
    if df[i].skew() < 0.5:
        print(f'{i} is left skewed\n')
    elif df[i].skew() > 0.5:
        print(f'{i} is Right skewed\n')
    else:
        print(f'{i} is symmetric\n')
```

```python
print("\nDATA CLEANING STARTED.........")

# Drop duplicates
df = df.drop_duplicates()

# Convert dates
for col in ["Order Date", "Ship Date"]:
    df[col] = pd.to_datetime(df[col], format="%m/%d/%Y", errors="coerce")

# Convert numeric columns
num_cols = ["Sales", "Quantity", "Discount", "Profit"]
for col in num_cols:
    df[col] = pd.to_numeric(df[col], errors="coerce")

# Fill numeric missing with median
df[num_cols] = df[num_cols].fillna(df[num_cols].median())

# Fill categorical missing with "Unknown"
df = df.fillna("Unknown")

print("\nCleaned Data Summary:")
df.info()
```

```python
# FEATURE ENGINEERING
df["Order_Year"] = df["Order Date"].dt.year
df["Order_Month"] = df["Order Date"].dt.month
df["Shipping_Delay_Days"] = (df["Ship Date"] - df["Order Date"]).dt.days.fillna(0)
```

```python
print("\n=== DATASET OVERVIEW ===")

print("\n--- Statistical Summary ---")
print(df.describe())

print("\n--- Categorical Column Overview ---")
for col in df.select_dtypes(include="object"):
    print(f"\nValue counts for {col}:")
    print(df[col].value_counts().head(5))
```

```python
def qq_plot_no_scipy(data, col):
    """Create QQ Plot manually using numpy"""
    data_sorted = np.sort(data)
    n = len(data_sorted)
    probs = (np.arange(1, n + 1) - 0.5) / n
    theoretical = np.quantile(np.random.normal(0, 1, n), probs)

    plt.figure(figsize=(7, 4))
    plt.scatter(theoretical, data_sorted, s=12)
    plt.plot(theoretical, theoretical, color="red")
    plt.title(f"QQ Plot of {col} (No SciPy)")
    plt.xlabel("Theoretical Quantiles")
    plt.ylabel("Sample Quantiles")
    plt.show()


def plot_all_distributions(col):
    data = df[col]
    print(f"\n========= {col} DISTRIBUTIONS =========")

    # Histogram
    plt.figure(figsize=(7,4))
    plt.hist(data, bins=40, edgecolor="black")
    plt.title(f"Histogram of {col}")
    plt.show()

    # KDE
    plt.figure(figsize=(7,4))
    sns.kdeplot(data, fill=True)
    plt.title(f"KDE Density of {col}")
    plt.show()

    # Boxplot
    sns.boxplot(x=data)
    plt.title(f"Boxplot of {col}")
    plt.show()

    # Violin Plot
    sns.violinplot(x=data)
    plt.title(f"Violin Plot of {col}")
    plt.show()

    # Rug Plot
    sns.rugplot(data, height=0.1)
    plt.title(f"Rug Plot of {col}")
    plt.show()

    # ECDF
    x = np.sort(data)
    y = np.arange(1, len(x)+1) / len(x)
    plt.figure(figsize=(7,4))
    plt.plot(x, y, marker=".", linestyle="none")
    plt.title(f"ECDF Plot of {col}")
    plt.xlabel(col)
    plt.ylabel("ECDF")
    plt.show()

    # QQ Plot (Custom)
    qq_plot_no_scipy(data, col)
```

```python
# Run distributions for numeric columns
for col in num_cols:
    plot_all_distributions(col)
```

```python
#  EXPLORATORY DATA ANALYSIS (EDA)

# Sales by Category
plt.figure(figsize=(8,5))
df.groupby("Category")["Sales"].sum().plot(kind="bar", color="skyblue")
plt.title("Sales by Category")
plt.ylabel("Total Sales")
plt.show()
```

```python
# Profit by Sub-Category
plt.figure(figsize=(12,5))
df.groupby("Sub-Category")["Profit"].sum().sort_values().plot(kind="bar", color="salmon")
plt.title("Profit by Sub-Category")
plt.ylabel("Total Profit")
plt.show()
```

```python
# Sales vs Profit Scatter
plt.figure(figsize=(8,5))
plt.scatter(df["Sales"], df["Profit"], alpha=0.4)
plt.title("Sales vs Profit")
plt.xlabel("Sales")
plt.ylabel("Profit")
plt.show()
```

```python
# Correlation Heatmap
plt.figure(figsize=(6,4))
sns.heatmap(df[num_cols].corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Pairplot
sns.pairplot(df[num_cols])
plt.show()
```
